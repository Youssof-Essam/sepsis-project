```python
lab_data = data["measurement_lab"].copy()
for column in lab_data.drop(["visit_occurrence_id","person_id","measurement_datetime","index"],axis = 1).columns:
    try:
        lab_data[column] = lab_data[column].apply(lambda x: DataWrangling.value_fixer(x,lab_extremes_sepsis_tolerant[column]["max"]))
    except:
        continue

lab_data["Blood arterial pH"] = lab_data["Blood arterial pH"].apply(lambda x: x + 6 if x < 6 else x)
lab_data["Blood arterial pH"] = lab_data["Blood arterial pH"].apply(lambda x: DataWrangling.value_fixer(x,1000,abs_max= 8))

lab_data["Blood venous pH"] = lab_data["Blood venous pH"].apply(lambda x: x + 6 if x < 6 else x)
lab_data["Blood venous pH"] = lab_data["Blood venous pH"].apply(lambda x: DataWrangling.value_fixer(x,1000,abs_max= 8))


lab_data
```

```python
lab_data_imputed = lab_data.copy()

for column in lab_data_imputed.columns:
    lab_data_imputed = DataWrangling.last_observed_imputer(lab_data_imputed,column)
lab_data_imputed
```

```python
nans = lab_data_imputed.isna().sum().sort_values(ascending = False) / lab_data_imputed.shape[0]
nans
```

```python
cols_to_drop = nans[nans > 0.26]
cols_to_drop.index.tolist()
```
