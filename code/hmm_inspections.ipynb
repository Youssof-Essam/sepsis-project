```python
import pandas as pd
import numpy as np
import joblib
from pathlib import Path
PROJECT_ROOT = Path("..").resolve()
PROJECT_ROOT
```

```python
hmm_all = joblib.load(f"{PROJECT_ROOT}/models/hmms/hmm_all.pkl")
all_data = pd.read_csv(f"{PROJECT_ROOT}/models/hmms/hmm_all.csv")

hmm_septic = joblib.load(f"{PROJECT_ROOT}/models/hmms/hmm_septic.pkl")
septic_data = pd.read_csv(f"{PROJECT_ROOT}/models/hmms/hmm_septic.csv")

hmm_balanced = joblib.load(f"{PROJECT_ROOT}/models/hmms/hmm_balanced.pkl")
balanced_data = pd.read_csv(f"{PROJECT_ROOT}/models/hmms/hmm_balanced.csv")
```

```python
lengths = all_data.groupby("person_id").size().tolist()
features = all_data.drop(["person_id", "measurement_datetime", "SepsisLabel"], axis=1).columns
X = all_data[features].to_numpy()

results = hmm_all.predict(X,lengths)
pd.Series(results).value_counts(normalize=True)
```

```python
final = all_data.copy()
final['state'] = pd.Series(results)
pd.get_dummies(final,columns=["SepsisLabel"],dtype=np.int8).groupby("state")[["SepsisLabel_0","SepsisLabel_1"]].mean()
```
